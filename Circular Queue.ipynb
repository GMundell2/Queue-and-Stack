* <font color = 'red'>MyCircularQueue(k)</font>: Constructor, set the size of the queue to be k.
* <font color = 'red'>Front</font>: Get the front item from the queue. If the queue is empty, return -1.
* <font color = 'red'>Rear</font>: Get the last item from the queue. If the queue is empty, return -1.
* <font color = 'red'>enQueue(value)</font>: Insert an element into the circular queue. Return true if the operation is successful.
* <font color = 'red'>deQueue()</font>: Delete an element from the circular queue. Return true if the operation is successful.
* <font color = 'red'>isEmpty()</font>: Checks whether the circular queue is empty or not.
* <font color = 'red'>isFull()</font>: Checks whether the circular queue is full or not.

```python
class MyCircularQueue:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.queue = [None for i in range(k)]
        self.head = -1
        self.tail = -1
        self.length = k

    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if self.isFull():
            return False
        else:
            if self.tail == -1:
                self.tail += 1
                self.head += 1
                self.queue[self.tail] = value
            elif self.tail < self.length - 1:
                self.tail += 1
                self.queue[self.tail] = value
            else:
                self.tail = 0
                self.queue[self.tail] = value
            return True

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if self.isEmpty():
            return False
        else:
            if self.head == self.tail:
                self.queue[self.head] = None
                self.head = -1
                self.tail = -1
            elif self.head < self.length - 1:
                self.queue[self.head] = None
                self.head += 1
            else:
                self.queue[self.head] = None
                self.head = 0
            return True  

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        if self.isEmpty():
            return -1
        else:
            return self.queue[self.head]

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        if self.isEmpty():
            return -1
        else:
            return self.queue[self.tail]

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        if self.head == self.tail == -1:
            return True
        else:
            return False

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        if self.head - self.tail == 1 or self.tail - self.head == self.length-1:
            return True
        else:
            return False
```
